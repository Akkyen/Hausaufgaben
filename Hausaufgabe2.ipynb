```python
import os
import time

import networkit as nk

files = list()

for file in os.listdir("./graphs"):
    if file.endswith(".graph"):
        T1_start = time.process_time()
        G: nk.Graph = nk.readGraph("./graphs/" + file, nk.Format.METIS)

        print()
        print(file)
        print(G)

        T2_start = time.process_time()
        diameter = nk.distance.Diameter(G, nk.distance.DiameterAlgo.Automatic, error=2.)
        diameter.run()
        print(diameter.getDiameter())
        T1_end = time.process_time()
        T2_end = time.process_time()

        print("Total processing time: ", (T1_end - T1_start))
        print("Processing time for diameter: ", (T2_end - T2_start))
```
