```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkit as nk
```

```python
G = nk.readGraph("./MIT8.edgelist", nk.Format.EdgeListTabZero)
```

```python
print(G.numberOfNodes(), G.numberOfEdges())
```

```python
# a)
cc = nk.components.ConnectedComponents(G)
cc.run()
print("Number of components:", cc.numberOfComponents())
C = cc.extractLargestConnectedComponent(G)
```

```python
# b)
dd = sorted(nk.centrality.DegreeCentrality(C).run().scores(), reverse=True)
plt.xscale("log")
plt.xlabel("degree")
plt.yscale("log")
plt.ylabel("number of nodes")
plt.plot(dd)
plt.show()
```

```python
try:
    import powerlaw
    fit = powerlaw.Fit(dd)
except ImportError:
    print ("Module powerlaw could not be loaded")
    
try:
    import powerlaw
    print("Steigung: (fit.alpha)", fit.alpha)
except ImportError:
      print ("Module powerlaw could not be loaded")
        
try:
    import powerlaw
    print(fit.distribution_compare('power_law','exponential'))
except ImportError:
      print ("Module powerlaw could not be loaded")
```

```python
import math
from collections import Counter, defaultdict
from networkit import distance
from itertools import islice
import numpy as np

valuePairs = list(tuple())

medians = defaultdict(int)
maxs = defaultdict(int)

sources = list(islice(C.iterNodes(), None, C.numberOfNodes(), 25))


spsp = distance.SPSP(C, sources)
spsp.run()

npDistances = np.asarray(spsp.getDistances())

for distance in npDistances:
    upDistance = distance[distance < 100000]
    upDistance = upDistance[upDistance > 1]

    medians[int(np.median(upDistance))] += 1
    maxs[int(np.max(upDistance))] += 1

print("maxs: ")
print(maxs)
print("medians: ")
print(medians)
```
